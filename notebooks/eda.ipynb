```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Analyse des variables

#### Pregnancies

```python
data["pregnancies"].value_counts()
```

```python
plt.boxplot(data["pregnancies"])
```

```python
data_preg = data[data["pregnancies"] < 13]
data_preg.head()
```

#### Glucose

```python
plt.hist(data_preg["glucose"])
```

```python
(data_preg["glucose"] < 1).sum()
```

```python
data_preg["glucose"] = data_preg["glucose"].replace(0, data_preg["glucose"].mean())
data_preg.head()
```

```python
(data_preg["glucose"] < 1).sum()
```

#### Diastolic

```python
plt.hist(data_preg["diastolic"])
```

```python
(data_preg["diastolic"] < 2).sum()
```

```python
data_preg["diastolic"] = data_preg["diastolic"].replace(0, data_preg["diastolic"].mean())
data_preg.head()
```

```python
plt.hist(data_preg["diastolic"])
```

#### Triceps

```python
data_preg["triceps"].describe()
```
